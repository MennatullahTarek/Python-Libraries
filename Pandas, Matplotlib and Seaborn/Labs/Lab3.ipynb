##<p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> 1. Total Charges :</p>

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isnull().sum()
```

```python
avg = df['TotalCharges'].mean()
avg
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(value = avg)
```

```python
df['TotalCharges'].isnull().sum()
```

##<p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> 2. General categorical col :</p>

```python
missing_val = df.isnull().sum()
missing_val
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True,ignore_index=True)
df.head()
```

```python
categorical_col = ['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity',
'OnlineBackup','DeviceProtection','TechSupport', 'StreamingTV','StreamingMovies','PaperlessBilling','PaymentMethod']
```

## <p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> One Hot Encoding :</p>

```python
pd.get_dummies(df[categorical_col]).head()
```

```python
df = pd.concat([df, pd.get_dummies(df[categorical_col])], axis=1)
df.info()
```

```python
df.drop(columns=categorical_col, inplace=True)
df.info()
```

```python
bool_columns = []
for i in range(df.shape[1]):
    if df.iloc[:, i].dtype == bool:
        bool_columns.append(df.columns[i])
```

```python
df[bool_columns] = df[bool_columns].astype(int)
df[bool_columns]
```

```python
plt.figure(figsize=(20, 10))

sns.heatmap(df[bool_columns].corr(), annot=True, fmt=".2f", linewidths=2, cmap="magma", cbar_kws={"shrink": 0.8})

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=45, va='top')

plt.title("Categorical Heatmap")
plt.tight_layout()
plt.show()
```

##<p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> 3. Contract :</p>

```python
df['Contract'].value_counts()
```

##<p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> Ordinal Encoding :</p>

```python
ordinal_encoding_mapping = {"Month-to-month" : 1, "One year" : 2, "Two year" :3}
df["Contract"] = df["Contract"].replace(ordinal_encoding_mapping)
df.head()
```

```python
df['Contract'] = pd.to_numeric(df['Contract'], errors='coerce')
```

```python
df.isnull().sum()
```

##<p style="color:#bf959f;font-family:newtimeroman;font-size:150%;text-align:center"> 4. Churn :</p>

```python
df['Churn'].value_counts()
```

```python
ordinal_encoding_mapping = {"No" : 0, "Yes" : 1}
df['Churn'] = df['Churn'].replace(ordinal_encoding_mapping)
df.head()
```

```python
df['Churn'] = pd.to_numeric(df['Churn'], errors='coerce')
```

```python
df['Churn'].isnull().sum()
```

```python
Churn = df.pop ('Churn')
```

```python
df.head()
```
